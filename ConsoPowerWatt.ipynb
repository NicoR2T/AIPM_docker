```python
from aipowermeter.deep_learning_power_measure.power_measure import experiment, parsers
import pandas as pd
from datetime import datetime, timedelta

driver = parsers.JsonParser("/media/profenpoche/data/testNicolas/docker_repo/consommation")
exp_result = experiment.ExpResults(driver)
df_power = pd.DataFrame(exp_result.get_curve("nvidia_draw_absolute"))
df_power.set_index("date", inplace=True)
df_power.set_index(pd.to_datetime(df_power.index, unit='s') + timedelta(hours=2), inplace=True)
df_power
```

```python
import numpy as np
def compute_power(data, timecodes):
    # The data must be a dataframe with time in index and timecodes must frame the power wished
    df = data.loc[(data.index >= timecodes[0]) & (data.index <= timecodes[1])]
    v = [(df.index.values[i+1] - df.index.values[i])/np.timedelta64(1, 's') * (df.iloc[i+1, 0] + df.iloc[i, 0]) / 2 for i in range(len(df)-1)]
    return sum(v)
```

```python
power = compute_power(df_power,timecodes)
print(f"Conso GPU : {round(power, 2)} (joules)/ {round(power/3600, 2)} (Wh)")
#power = compute_power(conso_wm,timecodes)
#print(f"Conso globale : {round(power, 2)} (joules)/ {round(power/3600, 2)} (Wh)")
```
